```python
import pandas as pd
import numpy as np
```

```python
df_re = pd.read_csv('../CSV Files/main.csv')
df_re.head()
```

```python
size_dict = df_re.groupby(['User_id']).size().to_dict()
df_train  = df_re.groupby(['User_id']).apply(lambda x : x.sort_values(by = 'Index', ascending = True).head(round(len(x)*0.8))).reset_index(drop = True)
df_test   = df_re.groupby(['User_id']).apply(lambda x : x.sort_values(by = 'Index', ascending = True).tail(round(len(x)*0.2))).reset_index(drop = True)
```

```python
df_train.to_csv('../CSV Files/train.csv', index=False)
df_test.to_csv('../CSV Files/test.csv', index=False)
```

```python
number_reviews_per_user_test = df_test.groupby('User_id')['Product_id'].count().sort_values(ascending=False)
```

```python
number_reviews_per_user_test
```

```python
num_reviews_df_test = pd.DataFrame(np.unique(number_reviews_per_user_test, return_counts = True)).transpose()
num_reviews_df_test.rename(columns={0: "Review_count", 1: "User_count"})
```
